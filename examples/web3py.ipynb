## Example of Getting Blockchain Info

```python
# get blocks
blocks = list(map(lambda x: w3.eth.getBlock(x), range(10000, 10050)))
```

```python
# get transactions
transactions = list(map(lambda x: w3.eth.getTransaction(x), blocks[10]['transactions']))
```

```python
df_blocks = pd.DataFrame(blocks)
df_blocks
```

```python
# some primary visualizations
df_blocks['timestamp'] = pd.to_datetime(df_blocks['timestamp'], unit='s')
df_blocks['gasUsed'] = df_blocks['gasUsed'].astype(int)
df_blocks['gasLimit'] = df_blocks['gasLimit'].astype(int)
df_blocks['size'] = df_blocks['size'].astype(int)
df_blocks['difficulty'] = df_blocks['difficulty'].astype(int)
df_blocks['totalDifficulty'] = df_blocks['totalDifficulty'].astype(int)

df_blocks.plot(x='timestamp', y=['gasUsed', 'gasLimit', 'size', 'difficulty', 'totalDifficulty'], subplots=True, figsize=(15, 15))

df_blocks.plot(x='timestamp', y=['gasUsed', 'gasLimit'], subplots=True, figsize=(15, 15))

df_blocks.plot(x='timestamp', y=['size', 'difficulty', 'totalDifficulty'], subplots=True, figsize=(15, 15))
```

```python
latest_block = w3.eth.getBlock('latest')
```

```python
# get transactions from latest block
latest_transactions = list(map(lambda x: w3.eth.getTransaction(x), latest_block['transactions']))
```

```python
# some preliary visualizations
df_latest_transactions = pd.DataFrame(latest_transactions)
df_latest_transactions
```

```python
df_latest_transactions['gas'] = df_latest_transactions['gas']
df_latest_transactions['gasPrice'] = df_latest_transactions['gasPrice']
df_latest_transactions['nonce'] = df_latest_transactions['nonce']
df_latest_transactions['value'] = df_latest_transactions['value']
df_latest_transactions['transactionIndex'] = df_latest_transactions['transactionIndex']

df_latest_transactions.plot(x='transactionIndex', y=['gas', 'gasPrice', 'nonce', 'value'], subplots=True, figsize=(15, 15))
```
