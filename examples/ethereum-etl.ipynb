```python
# basic visualizations

import matplotlib.pyplot as plt

plt.figure(figsize=(12, 8))
plt.hist(blocks['gas_used'], bins=100)
plt.title('Gas Used Distribution')
plt.show()

plt.figure(figsize=(12, 8))
plt.hist(blocks['gas_limit'], bins=100)
plt.title('Gas Limit Distribution')
plt.show()

plt.figure(figsize=(12, 8))
plt.hist(blocks['difficulty'], bins=100)
plt.title('Difficulty Distribution')
plt.show()
```
