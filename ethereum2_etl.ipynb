# Part IV How to Merge the Beacon Chain and the Execution Layer Data?

(answer: by block hash)


*  [etherscan](https://etherscan.io/block/0xd049fc79f832e3a4eeb0465c8681c48cc737a047a88b5cab491d703e02e05be4)
*  [beaconscan](https://beaconscan.com/slot/6346591)

for the above state, we find from etherscan and beaconscan that


*   when slot number =4700013, the block number=15533942
*   when slot number = 4701013, the block number =15535581


Now we can query the connected execution layer data from etherem-etl

https://github.com/blockchain-etl/ethereum-etl
(currently only works with Python<=3.9)

```python
!ethereumetl export_blocks_and_transactions --start-block 15533942 --end-block 15535581 --provider-uri https://multi-restless-replica.quiknode.pro/a5255e1386c2c5579770407e25b8f32f3be811cc/ --blocks-output blocks.csv
```

```python
blocks = pd.read_csv("https://raw.githubusercontent.com/sunshineluyao/eth-data-collection/main/data/blocks.csv")
blocks.head()
```

```python
blocks= blocks.sort_values(by=["number"])
```

```python
df_slot = df_beacon[["block_slot","block_epoch","eth1_block_hash"]]
df_slot
```

```python
df_block = blocks[["number","hash"]]
df_block.head()
```

```python
slot_block=df_slot.merge(df_block, left_on='eth1_block_hash', right_on='hash')
slot_block
```

```python
slot_block.to_csv("slot_block.csv")
```
