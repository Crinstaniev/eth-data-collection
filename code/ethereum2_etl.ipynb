# Part III Process and Visualize Data

```python
df_beacon["skipped"].value_counts()
```

```python
# basic stats
df_beacon.info()
```

```python
df_beacon= df_beacon.sort_values(by=["block_slot"])
```

```python
from plotly.subplots import make_subplots
import plotly.graph_objects as go
import plotly.io as pio
layout = go.Layout(
    autosize=False,
    width=1800,
    height=1800
)

fig = make_subplots(
    rows=2, cols=1,
    subplot_titles=("block_epoch", "eth1_deposit_count"))

fig.add_trace(go.Scatter(x=df_beacon.block_slot, y=df_beacon.block_epoch),
              row=1, col=1)

fig.add_trace(go.Scatter(x=df_beacon.block_slot, y=df_beacon.eth1_deposit_count),
              row=2, col=1)

fig.update_layout(height=900, width=900,
                  title_text="Etherem Beacon Chain Deposit Count Visualization via Etherem2-ETL")

fig.show()
```

```python
pio.write_image(fig, 'ETL2_deposit.png')
```
