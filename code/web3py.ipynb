# Part III Process and Visualize the Blockchain Block and Transaction Data

```python
df_blocks.dtypes
```

```python
# Process the data for visualization
df_blocks['timestamp'] = pd.to_datetime(df_blocks['timestamp'], unit='s')
df_blocks['gasUsed'] = df_blocks['gasUsed'].astype(int)
df_blocks['gasLimit'] = df_blocks['gasLimit'].astype(int)
df_blocks['size'] = df_blocks['size'].astype(int)
df_blocks['difficulty'] = df_blocks['difficulty'].astype(int)
df_blocks['totalDifficulty'] = pd.to_numeric(df_blocks['totalDifficulty'],errors='coerce')
```

```python
df_blocks['timestamp']
```

```python
df_blocks.columns
```

```python
from plotly.subplots import make_subplots
import plotly.graph_objects as go
import plotly.io as pio
layout = go.Layout(
    autosize=False,
    width=1800,
    height=1800
)

fig = make_subplots(
    rows=2, cols=2,
    subplot_titles=("baseFeePerGas", "gasLimit", "gasUsed", "size"))

fig.add_trace(go.Scatter(x=df_blocks.timestamp, y=df_blocks.baseFeePerGas),
              row=1, col=1)

fig.add_trace(go.Scatter(x=df_blocks.timestamp, y=df_blocks.gasLimit),
              row=1, col=2)

fig.add_trace(go.Scatter(x=df_blocks.timestamp, y=df_blocks.gasUsed),
              row=2, col=1)

fig.add_trace(go.Scatter(x=df_blocks.timestamp, y=df_blocks['size']),
              row=2, col=2)

fig.update_layout(height=900, width=900,
                  title_text="Ethereum Blockchain Block Data Visualization via Web3py")

fig.show()
```

```python
pio.write_image(fig, 'Web3py_blocks.png')
```
